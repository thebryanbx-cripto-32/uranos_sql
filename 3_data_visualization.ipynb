# Data Visualization
### Loading the data

```python
import pandas as pd
import matplotlib.pyplot as plt
wines = pd.read_csv("data/wines.csv", sep=';')
acidity = pd.read_csv("data/acidity.csv", sep=';')
composition = pd.read_csv("data/composition.csv", sep=';')
#df2 = pd.read_csv("sql_result/result")
```

### Adding data from the result table.

```python
df_avg = pd.DataFrame({
    "quality": [3,4,5,6,7,8,9],
    "avg_alcohol": [10.21,10.22,9.87,10.65,11.51,11.91,12.18],
    "avg_density": [0.9957,0.9948,0.9957,0.9944,0.9928,0.9921,0.9915],
    "avg_volatile_acidity": [0.517,0.462,0.394,0.316,0.292,0.303,0.298],
    "avg_residual_sugar": [5.14,4.03,5.48,5.15,4.17,4.77,4.12]})

df_avg
```

### H1: Alcohol vs Quality
Higher alcohol levels are associated with higher quality.



```python
plt.figure()
plt.bar(df_avg["quality"], df_avg["avg_alcohol"])
plt.xlabel("Quality")
plt.ylabel("Average Alcohol")
plt.title("Average Alcohol by Wine Quality")
plt.ylim(6, 13)    
plt.show()
```

### H2: Desnity vs Quality
Density tends to decrease as quality increases.

```python
plt.style.use("default")
plt.figure()
plt.bar(df_avg["quality"], df_avg["avg_density"])
plt.xlabel("Quality")
plt.ylabel("Average Density")
plt.title("Average Density by Wine Quality")
plt.ylim(0.991, 0.996)
plt.show()
```

## H3: Volatile Acidity vs quality
Higher volatile acidity is linked to lower quality.

```python
plt.figure()
plt.bar(df_avg["quality"], df_avg["avg_volatile_acidity"])
plt.xlabel("Quality")
plt.ylabel("Avg Volatile Acidity")
plt.title("Volatile Acidity by Wine Quality")
plt.show()
```

## H4: Residual Sugar vs Quality
Residual sugar shows no clear linear pattern.

```python
plt.figure()
plt.bar(df_avg["quality"], df_avg["avg_residual_sugar"])
plt.xlabel("Quality")
plt.ylabel("Avg Residual Sugar")
plt.title("Residual Sugar by Wine Quality")
plt.ylim(2, 6)
plt.show()
```

## Distribution of Wine Quality

```python
quality = [3, 4, 5, 6, 7, 8, 9]
total_wines = [30, 206, 1751, 2323, 855, 148, 5]

plt.figure()
plt.bar(quality, total_wines)

plt.xlabel("Quality")
plt.ylabel("Number of Wines")
plt.title("Distribution of Wine Quality")
plt.ylim(0, 2800) 

plt.show()
```
